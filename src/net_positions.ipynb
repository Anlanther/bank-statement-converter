## Extras

```python
sc_statement_path = 'Bank.SC.csv'
citi_statement_path = 'Bank.CITI.csv'
hsbc_one_statement_path = 'Bank.HSBC_ONE.csv'
```

```python
sc_df = pd.read_csv(sc_statement_path).rename(columns={'amount': 'sc_amount'})
citi_df = pd.read_csv(citi_statement_path).rename(columns={'amount': 'citi_amount'})
hsbc_one_df = pd.read_csv(hsbc_one_statement_path).rename(columns={'amount': 'hsbc_one_amount'})
```

```python
combined_df = sc_df.merge(citi_df, on='date', how='outer')\
                   .merge(hsbc_one_df, on='date', how='outer')\
                   .sort_values('date').reset_index(drop=True)\
                   .fillna(0)
combined_df['total'] = combined_df['sc_amount'] + combined_df['citi_amount'] + combined_df['hsbc_one_amount']

combined_df
```

```python
combined_df.to_csv(f'{os.getcwd()}/combined.csv', index=False, mode='w+')
```
